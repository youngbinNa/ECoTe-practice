나동빈 답

```python
n=7
t=[3,5,1,1,2,1,1]
p=[10,20,10,20,15,40,10]
array=[(3,10),(5,20),(1,10),(1,20),(2,15),(1,40),(1,10)]
#array=[]
#for i in range(n):
#  array.append(list(map(int,input().split())))
#array=[(1,1),(1,2),(1,3),(1,4),(1,5),(1,6),(1,7),(1,8),(1,9),(1,10)]
max_pay=-1e9
# 리스트를 뒤에서부터 거꾸로 확인
dp = [0] * (n + 1) # 다이나믹 프로그래밍을 위한 1차원 DP 테이블 초기화
max_value = 0
for i in range(n - 1, -1, -1):
    time = t[i] + i # t[i]: 상담기간, i:인덱스 (상담시간+인덱스(현재위치)=time이 리스트 끝을 넘으면 안되니까)
    print('i:',i,'t[i]:',t[i],'time:',time,'pay:',p[i])
    # 상담이 기간 안에 끝나는 경우
    if time <= n:
        # 점화식에 맞게, 현재까지의 최고 이익 계산
        print('max_value:',max_value,'dp[i]:',dp[i],'dp[time]:',dp[time])
        #뒤에서 부터 계산, (2,15)까지 왔을때, p[i]=15,dp[time]=10(이틀 뒤 인덱스의 금액), 35(현재금액+이틀 뒤 금액)보다 건너뛰고 (1,40)의 40이 금액 더 커서 
        dp[i] = max(p[i] + dp[time], max_value)
        max_value = dp[i]
    # 상담이 기간을 벗어나는 경우
    else:
        dp[i] = max_value
    print(dp)
    print()

print(max_value)
```
