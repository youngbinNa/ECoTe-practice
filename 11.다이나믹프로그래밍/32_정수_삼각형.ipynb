나동빈 답

```python
#n=int(input())
#array=[[0]*(i+1) for i in range(n)]
n=5
array=[[7], [3, 8], [8, 1, 0], [2, 7, 4, 4], [4, 5, 2, 6, 5]]
#array=[]
#for i in range(5):
#  array.append(list(map(int,input().split())))
#print(array)

for i in range(1,n):
  for j in range(len(array[i])):
    if j==0:
      left=0
    else:
      left=array[i-1][j-1]
    if j==len(array[i])-1:
      right=0
    else:
      right=array[i-1][j] 
    array[i][j]=array[i][j]+max(left,right)
print(max(array[n-1]))
```
