나동빈 답

```python
'''
n*m크기의 금광 (1*1크기의 칸)
맨 처음에는 첫 번째 열의 어느 행에서든 출발
이후 m-1번에 걸쳐 매번 오른쪽 위,오른쪽,오른쪽 아래 3가지 중 하나의 위치로 이동
채굴자가 얻을 수 있는 금의 최대 크기
1  3 3 2
② 1 4 1
0 ⑥ ④ ⑦ 최대크기 19
첫째 줄에 테스트 케이스 T 입력 (1<=T<=1,000)
매 테스트 케이스 첫째 줄에 n과 m이 공백으로 구분되어 입력 (1<=n,m<=20)
둘째 줄에 n*m개의 위치에 매장된 금의 개수가 공백으로 구분되어 입력 (1<=각 위치에 매장된 금의 개수<=100
입력
2
3 4
1 3 3 2 2 1 4 1 0 6 4 7
4 4
1 3 5 2 2 4 1 5 0 2 3 0 6 1 2
출력
19
16
'''
#테스트 케이스 입력
#for tc in range(int(input())):
for tc in range(1):
  #금광 정보 입력
  #n,m = map(int,input().split())
  n,m=3,4
  #array=list(map(int,input().split()))
  array=[1, 3, 3, 2, 2, 1, 4, 1, 0, 6, 4, 7]
  #다이나믹 프로그래밍을 위한 2차원 dp 테이블 초기화
  dp=[]
  index=0
  for i in range(n):
    dp.append(array[index:index+m])
    index += m
  #다이나믹 프로그래밍 진행
  for j in range(1,m):
    for i in range(n):
      print(i,',',j)
      #왼쪽 위에서 오는 경우
      if i==0: 
        left_up=0
      else: 
        left_up=dp[i-1][j-1]
        print('left_up:',left_up)
      #왼쪽 아래에서 오는 경우
      if i==n-1: 
        left_down=0
      else: 
        left_down=dp[i+1][j-1]
        print('left_down:',left_down)
      #왼쪽에서 오는 경우
      left=dp[i][j-1]
      print('left:',left)
      dp[i][j]=dp[i][j]+max(left_up,left_down,left)
      '''
      0 , 1
      left_down: 2
      left: 1
      1 , 1
      left_up: 1
      left_down: 0
      left: 2
      2 , 1
      left_up: 2
      left: 0
      0 , 2
      left_down: 3
      left: 5
      1 , 2
      left_up: 5
      left_down: 8
      left: 3
      2 , 2
      left_up: 3
      left: 8
      0 , 3
      left_down: 12
      left: 8
      1 , 3
      left_up: 8
      left_down: 12
      left: 12
      2 , 3
      left_up: 12
      left: 12
      '''
  result=0
  print(dp)
  #[[1, 5, 8, 14], [2, 3, 12, 13], [0, 8, 12, 19]]
  for i in range(n):
    result=max(result, dp[i][m-1]) # 14 13 19 중 큰수 result
  print(result)
```
