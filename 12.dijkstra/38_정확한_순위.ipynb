```python
INF=1000
n,m=6,6

graph=[[INF]*(n+1) for i in range(n+1)]

for a in range(1,n+1):
  for b in range(1,n+1):
    if a==b:
      graph[a][b]=0
'''
for _ in range(m):
  a,b=map(int,input().split())
  #비용 1로 설정
  graph[a][b]=1
print(graph)
'''
graph=[[1000, 1000, 1000, 1000, 1000, 1000, 1000], [1000, 0, 1000, 1000, 1000, 1, 1000], [1000, 1000, 0, 1000, 1000, 1000, 1000], [1000, 1000, 1000, 0, 1, 1000, 1000], [1000, 1000, 1, 1000, 0, 1000, 1], [1000, 1000, 1, 1000, 1, 0, 1000], [1000, 1000, 1000, 1000, 1000, 1000, 0]]

for k in range(1, n + 1):
  for a in range(1, n + 1):
    for b in range(1, n + 1):
      graph[a][b] = min(graph[a][b], graph[a][k] + graph[k][b])
      
for a in range(1, n + 1):
  for b in range(1, n + 1):
    if graph[a][b]==INF:
      print('INF',end=" ")
    else:
      print(graph[a][b],end=" ")
  print()
  
result=0  
for a in range(1, n + 1):
  count=0
  for b in range(1, n + 1):
    #양방향 모두 갈 수 있으면
    if graph[a][b]!=INF or graph[b][a]!=INF:
      print(a,b,graph[a][b],graph[b][a],count)
      count+=1
  # n(6)개 전부 갈 수 있으면 성적 순위 다 알 수 있다는 뜻
  if count==n:
    print(a,b)
    result+=1
  print()
print(result)
```
