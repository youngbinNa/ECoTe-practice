```python
n=6
data=[1,2,3,4,5,6]
add,sub,mul,div=2,1,1,1 # + - * /

min_value=1e9
max_value=-1e9

def dfs(i,now):
  global add,sub,mul,div,min_value,max_value
  print('i:',i,'now:',now)
  if i==n:
    print('min:',min_value,'max:',max_value,'i:',i,'now:',now)
    min_value=min(min_value,now)
    max_value=max(max_value,now)
    print('min:',min_value,'max:',max_value,'i:',i,'now:',now)
  else:
    if add>0:
      print('in add now:',now,'data[i]:',data[i],'i:',i)
      add-=1
      dfs(i+1,now+data[i])
      add+=1
      print('in add now:',add,mul)
    if sub > 0:
      sub -= 1
      dfs(i + 1, now - data[i])
      sub += 1
    if mul > 0:
      print('in mul now:',now,'data[i]:',data[i],'i:',i)
      mul -= 1
      dfs(i + 1, now * data[i])
      mul += 1
      print('in mul now:',add,mul)
    if div > 0:
      div -= 1
      dfs(i + 1, int(now / data[i])) # 나눌 때는 나머지를 제거
      div += 1
    
dfs(1,data[0])
print(max_value)
print(min_value)
```
